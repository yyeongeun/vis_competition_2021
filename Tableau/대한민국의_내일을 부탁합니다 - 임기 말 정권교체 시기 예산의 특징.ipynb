### 2.3. 박근혜 정부-문재인 정부 주요 복지정책 예산 비교
 2007년 서브프라임 모기지론 사태가 발생했습니다.  치솟던 미국의 자산 가격에 편승하여 저신용자(sub-prime)의 주택담보대출이 무분별하게 실행되었습니다. 미국의 경상수지와 서비스 수지가 모두 적자로 돌아 섰고, 시장에 풀린 저신용자의 대출금을 회수하지 못한 금융기관들이 파산하기 시작합니다. 이러한 상황은 우리나라에도 영향을 끼쳐 부동산 시장이 얼어붙는 결과를 낳게 됩니다. 
 금리 인상으로 인해 대출로 집을 구매한 사람들은 수입의 대부분을 이자로 내는 “ 하우스푸어” 로 전락하는 상황에서 박근혜 정부는 어떻게 대처 하였을까요? 정부가 가계와 기업의 위기를 바라보는 시각을 양 정부의 예산 변화를 통해 분석해봅니다.
